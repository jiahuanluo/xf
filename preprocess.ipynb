```python
train_df = pd.read_csv('data/train/train.txt', header=None, index_col=None)
test_df = pd.read_csv('data/test/apply_new.txt', header=None, index_col=None)
```

```python
train_df.columns = ['pid', 'label', 'gender', 'age', 'tagid', 'time','province', 'city', 'maker', 'model']
test_df.columns = ['pid', 'gender', 'age', 'tagid', 'time','province', 'city', 'maker', 'model']
print(train_df.head(3))
print(test_df.head(3))
```

```python
train_df['time'] = train_df['time'].apply(lambda x: x.strip().replace('[', '').replace(']', '').split(',')).apply(lambda x: list(map(float, x)))
test_df['time'] = test_df['time'].apply(lambda x: x.strip().replace('[', '').replace(']', '').split(',')).apply(lambda x: list(map(float, x)))
train_df['tagid'] = train_df['tagid'].apply(lambda x: x.strip().replace('[', '').replace(']', '').split(',')).apply(lambda x: list(map(int, x)))
test_df['tagid'] = test_df['tagid'].apply(lambda x: x.strip().replace('[', '').replace(']', '').split(',')).apply(lambda x: list(map(int, x)))
```

```python
sorted_tag_list = []
sorted_time_list = []
for tag_list, time_list in train_df[['tagid', 'time']].values:
    tag_time_dict = {}
    for tag, time in zip(tag_list, time_list):
        tag_time_dict[tag] = time
    tag_time = sorted(tag_time_dict.items(), key = lambda x: x[1])
    sorted_tag_list.append([t_t[0] for t_t in tag_time])
    sorted_time_list.append([t_t[1] for t_t in tag_time])
train_df['sorted_tagid'] = sorted_tag_list
train_df['sorted_time'] = sorted_time_list
```

```python
sorted_tag_list = []
sorted_time_list = []
for tag_list, time_list in test_df[['tagid', 'time']].values:
    tag_time_dict = {}
    for tag, time in zip(tag_list, time_list):
        tag_time_dict[tag] = time
    tag_time = sorted(tag_time_dict.items(), key = lambda x: x[1])
    sorted_tag_list.append([t_t[0] for t_t in tag_time])
    sorted_time_list.append([t_t[1] for t_t in tag_time])
test_df['sorted_tagid'] = sorted_tag_list
test_df['sorted_time'] = sorted_time_list
```

```python
data = pd.concat([train_df, test_df])
data
```

```python
data["gender"] = data["gender"].fillna(-1).astype("int")
data["age"] = data["age"].fillna(-1).astype("int")
data["province"] = data["province"].fillna("未知")
data["city"] = data["city"].fillna("未知")
data["model"] = data["model"].fillna("未知")
data["maker"] = data["maker"].fillna("未知")
data
```

```python
from sklearn.preprocessing import LabelEncoder
sparse_feats = ['gender', 'age', 'province', 'city', 'model', 'maker']
label_encoder_dict = {}
for feat in sparse_feats:
    label_encoder_dict[feat] = LabelEncoder()
    label_encoder_dict[feat].fit(list(set(data[feat])))
    data[feat] = label_encoder_dict[feat].transform(data[feat])
data
```

```python
data.to_csv("./data/tran_test_transform.csv", index=None)
train_df.to_csv("./data/train/train_transform.csv", index=None)
test_df.to_csv("./data/test/test_transform.csv", index=None)
```
